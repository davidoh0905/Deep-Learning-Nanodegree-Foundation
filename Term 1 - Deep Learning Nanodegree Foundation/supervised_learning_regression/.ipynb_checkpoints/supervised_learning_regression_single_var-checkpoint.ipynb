# Linear Regression
- body_reg variable is a linear regression model that has been fitted to the data x_values and y_values
- Fitting the model means finding the best line that fits the training data.

```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt

#read data
dataframe = pd.read_fwf('brain_body.txt')
x_values = dataframe[['Brain']]
y_values = dataframe[['Body']]

#train model on data
body_reg = linear_model.LinearRegression()
### scikit-learn's LinearRegression class provides fit function.
body_reg.fit(x_values, y_values)

#visualize results
plt.scatter(x_values, y_values)
plt.show()
```

```python
plt.plot(x_values, body_reg.predict(x_values))
plt.show()
```

```python
plt.scatter(x_values, y_values)
plt.plot(x_values, body_reg.predict(x_values))
plt.show()
```

# Making Predictions
- use predict() function
- the reason for predicting on an array like [127] not just 127 is because you can have a model that makes a prediction using multiple features
- we'll go over using multiple variables in linear regression later

```python
print(body_reg.predict([[127],[248]]))
```

# Regression Warning!
- Linear regression comes with a set of implicit assumptions and is not the best model for every situation

## Linear Regression Works Best when the data is linear
- linear regression produces a straight line model from the training data.
- if the relationship in the training data is not really linear, you will need to either make adjustments(transform your training data), add features, or use another model

## Linear Regression is sensitive to outliers
- Linear regression tries to find a 'best fit' line among the training data
- if your dataset has some outliers that don't fit a general pattern, they can have surprisingly large effect
